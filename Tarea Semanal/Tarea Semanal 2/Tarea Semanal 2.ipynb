##### Pasa Bajos en Python

##### Pasa Banda en Python

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sat Apr  8 13:48:47 2023

@author: nahu
"""

# Librerías externas NumPy, SciPy y Matplotlib
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import numpy as np

from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot
# Componentes Utilizados
R1 = 1e3
R2 = 30e3
R3 = R4 = 10e3
C = 100e-6

#Cálculo de wo y q
wo = 1/(C*R3)
Q = C*R2*wo
plt.close('all')
    
Tf = TransferFunction( [-1/(C*R1), 0], [1, wo/Q, wo**2] )

bodePlot(Tf, fig_id=1 )

pzmap(Tf, fig_id=2) 

GroupDelay(Tf, fig_id=3)
```

##### Conclusiones
Debido a que el cálculo analítico y la simulación en python se realiza en frecuencia angular (rad/seg) y la simulación en Spice se realiza en frecuencia (1/seg o Hz), se debe tener en cuenta que wo = 1 implica una frecuencia de 160 mHz aproximadamente. Teniendo en cuenta dicha salvedad, se observa que ambas simulaciones coinciden con lo bosquejado analíticamente. 
En el caso del pasabajos, la ganancia de continua resulta efectivamente 20 dB, mientras que en el pasa altos la ganancia en la banda de paso es de 30 veces, o lo que es lo mismo para este caso, 30 dB.
Por otro lado, en ambos casos se oberva un sobrepico en la transferencia.

##### Pasa Bajos Butterworth en Python

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sat Apr  8 13:48:47 2023

@author: nahu
"""

# Librerías externas NumPy, SciPy y Matplotlib
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import numpy as np

from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot
# Componentes Utilizados
R1 = 1e3
R2 = 21.213e3
R3 = R4 = 10e3
C = 100e-6

#Cálculo de wo y q
wo = 1/(C*R3)
Q = C*R2*wo
plt.close('all')
    
Tf = TransferFunction( [-wo/(C*R1)], [1, wo/Q, wo**2] )

bodePlot(Tf, fig_id=1 )

pzmap(Tf, fig_id=2) 

GroupDelay(Tf, fig_id=3)
```

##### Conclusiones
Como el cálculo analítico y la simulación en python se realiza en frecuencia angular (rad/seg) y la simulación en Spice se realiza en frecuencia (1/seg o Hz), se debe tener en cuenta que wo = 1 implica una frecuencia de 160 mHz aproximadamente. 
Teniendo en cuenta dicha aclaración, se observa que ambas simulaciones coinciden. Además, se redujo un poco el sobrepico respecto a la transferencia pasa bajos que no es Butterworth. Esto se debe a que se redujo el valor de Q.
