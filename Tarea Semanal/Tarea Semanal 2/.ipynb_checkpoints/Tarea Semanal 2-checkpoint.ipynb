## Simulación Numérica en Python 

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sat Apr  8 13:48:47 2023

@author: nahu
"""

# Librerías externas NumPy, SciPy y Matplotlib
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import numpy as np

from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot
# Componentes Utilizados
R1 = 1/10
R2 = 3
R3 = R4 = 1
C = 1

#Cálculo de wo, q y K
wo = 1/(C*R3)
Q = C*R2*wo
K = R3/R1
plt.close('all')
    
Tf = TransferFunction( [-K], [1, wo/Q, wo**2] )

bodePlot(Tf, fig_id=1)

pzmap(Tf, fig_id=2) 

GroupDelay(Tf, fig_id=3)
```

## Simulación Circuital en LT Spice

A continuación, se adjunta el circuito simulado



Simule directamente con la red normalizada, asignandole los valores de K=10 y Q=3, por que lo pedía la consigna. Coincide logicamente con lo calculado analiticamente ANTES de normalizar la red. Por otro lado, el capacitor vale 1/2pi para que el punto de interés (valor de la transferencia en wo) quede en 1 Hz. Es como si hubiera cambiado la norma de frecuencia, en lugar de wo=1, wo=2pi. 

El módulo de la transferencia y la respuesta en fase que se obtuvieron a partir de la simulación se presentan a continuación.






## Conclusiones sobre el Pasa Bajos Ackerberg Mosberg
De acuerdo a lo analizado analíticamente, para wo = 1 se espera una ganancia de K.Q. De acuerdo a los valores propuestos por el enunciado, la ganancia debería ser de 30 veces, o lo que es lo mismo aproximadamente 29,54 db. Se observa que en ambas simulaciones dicho valor coincide con el esperado.
Además, según lo analizado analíticamente, para wo = 0 se espera una ganancia de K. De acuerdo a los valores propuestos por el enunciado, la ganancia debería ser de 10 veces, o lo que es lo mismo 20 db.
Se observa que en ambas simulaciones dicho valor coincide con el esperado.

# Pasa Bajos Ackerberg Mosberg Butterworth

## Cálculo Analítico
A continuación, se adjunta el cálculo analítico requerido para lograr que la transferencia de la red corresponda a la de un Butter de orden 2.

```python
from IPython.display import IFrame
IFrame("./Butter.pdf", width=900, height=600)
```

## Simulación Numérica en Python

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sat Apr  8 13:48:47 2023

@author: nahu
"""

# Librerías externas NumPy, SciPy y Matplotlib
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import numpy as np
import math

from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot
# Componentes Utilizados
R1 = 1/10
R2 = math.sqrt(2)/2
R3 = R4 = 1
C = 1

#Cálculo de wo, q y K
wo = 1/(C*R3)
Q = C*R2*wo
K = R3/R1
plt.close('all')
    
Tf = TransferFunction( [-K], [1, wo/Q, wo**2] )

bodePlot(Tf, fig_id=1)

pzmap(Tf, fig_id=2) 

GroupDelay(Tf, fig_id=3)
```

## Simulación Circuital en Lt Spice
El circuito Simulado en Lt Spice se adjunta a continuación. Lo único que cambia respecto al anterior es el valor de Q, y por consecuencia el valor de R2.



El módulo de la transferencia y la respuesta en fase que se obtuvieron a partir de la simulación se presentan a continuación.





## Conclusiones sobre el Pasa Bajos Butterworth Ackerberg Mosberg
De acuerdo a lo analizado analíticamente, para wo = 1 se espera una ganancia de K.Q. Como ahora la transferencia es Butterworth, es decir que Q = sqrt(2)/2 y como se mantuvo el valor de K = 10, la ganancia debería ser de 7,07 veces, o lo que es lo mismo aproximadamente 16,98 db. Se observa que en ambas simulaciones dicho valor coincide con el esperado.
Además, según lo analizado analíticamente, para wo = 0 se espera una ganancia de K. De acuerdo a los valores propuestos por el enunciado, la ganancia debería ser de 10 veces, o lo que es lo mismo 20 db.
Se observa que en ambas simulaciones dicho valor coincide con el esperado.

Nota: La simulación en Spyder muestra un valor de 17 db para wo = 1, pero no encontré la manera de añadir un cursor como hice en Spice para poder mostrarlo.

## Simulación Numérica en Python

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sat Apr  8 13:48:47 2023

@author: nahu
"""

# Librerías externas NumPy, SciPy y Matplotlib
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import numpy as np

from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot
# Componentes Utilizados
R1 = 1/10
R2 = 3
R3 = R4 = 1
C = 1

#Cálculo de wo, q y k
wo = 1/(C*R3)
Q = C*R2*wo
K = R3/R1
plt.close('all')
    
Tf = TransferFunction( [-K, 0], [1, wo/Q, wo**2] )

bodePlot(Tf, fig_id=1 )

pzmap(Tf, fig_id=2) 

GroupDelay(Tf, fig_id=3)
```

Comparando ambas simulaciones, en Spice para 100 mHz la ganancia es de 0 db, y en python dicha ganancia también es de 0db, es decir que ambas simulaciones coinciden.
