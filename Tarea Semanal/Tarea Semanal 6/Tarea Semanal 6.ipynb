## Simulación Numérica

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
# -------------------------- Librerias ------------------------------
import sympy as sp

from pytc2.cuadripolos import calc_MAI_impedance_ij, calc_MAI_vtransf_ij_mn, calc_MAI_ztransf_ij_mn
from pytc2.general import print_latex

#--------------------------- Definicion de Variables Simbólicas ----------------------

S = sp.symbols('S', complex=True)
L1,C2,L3,R = sp.symbols('L1 C2 L3 R', real=True, positive=True)

#--------------------------- Armado de la MAI ----------------------------------

#               Nodos: 1                         2                        3               4
Ymai = sp.Matrix([  
                    [ 1/(S*L1),     -1/(S*L1),                                 0,               0],
    
                    [ -1/(S*L1),    (1/S)*(1/L1 + 1/L3) + S*C2,        -1/(S*L3),           -S*C2],
    
                    [ 0,            -1/(S*L3),                    1/(S*L3) + 1/R,            -1/R],
                    
                    [ 0,            -S*C2,                                  -1/R,      S*C2 + 1/R]
                 ])

#--------------------------- Cálculo de la Transferencia de Tensión -------------------------

con_detalles = False

print('Transferencia de tensión:')
Vmai = calc_MAI_vtransf_ij_mn(Ymai, 2, 3, 0, 3, verbose=con_detalles) 

print_latex( r'T^{{ {:d}{:d} }}_{{ {:d}{:d} }} = '.format(2, 3, 0, 3) + sp.latex(Vmai) )

print('')


#--------------------------- Cálculo de la Impedancia de Entrada -------------------------

print('Impedancia de Entrada:')
Zmai = calc_MAI_impedance_ij(Ymai, 0, 3, verbose=con_detalles)

print_latex( r'Z^{{ {:d}{:d} }} = '.format(0, 3, 0, 1) + sp.latex(Zmai) )

```

Finalmente, a partir de la MAI, se computa la transferencia de tensión y la impedancia de entrada del sistema. Se observa que la transferencia de tensión obtenida coincide con la calculada analíticamente.
